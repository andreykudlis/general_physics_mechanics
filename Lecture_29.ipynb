# Деформации. Волны

## Закон Гука
### Основные определения

<br>

**Деформация** - изменение формы/объёма тела под действием приложенных внешних сил. Она бывает двух видов: **упругой** (исчезает, если силы = 0) и **неупругой (пластичной)**.

Малые деформации, как правило, упругие. Максимальное значение прилагаемых сил определяется пределом упругости.

Рассмотрим растяжение стержня.

<br>

\begin{align}
    \text{Напряжение:}\;\sigma=\dfrac{F}{S}.
\end{align}
Если сила направлена _из_ тела, то она будет называться **натяжением** и обозначаться $\vec{T}$. Если _внутрь_ тела, то **давлением** $\vec{P}$.

<br>

\begin{align}
    \text{Относительное удлинение}\;\varepsilon = \dfrac{\Delta l}{l}.
\end{align}

<br>

### Закон Гука

<br>

**Закон Гука**: малые упругие деформации пропорциональны силам, которые их вызывают:
\begin{align}
    \varepsilon = \dfrac{\sigma}{E},
\end{align}

<br>

где $E$ - модуль Юнга.

Мы помним:

<br>

\begin{align}
    &F = kx\;\equiv\;k\Delta l\nonumber\\
    &\dfrac{\Delta l}{l} = \dfrac{\sigma}{E} = \dfrac{F}{ES} \implies\nonumber\\
    &\implies\boxed{k = \dfrac{ES}{l}}.
\end{align}

<br>

### Плотность упругой деформации

Энергия упругой деформации формируется за счёт работы силы по растяжению (сжатию) стержня:

<br>

\begin{align}
    U = \int^{\Delta l}_{0} F(x)\mathrm{d}x = \int^{\Delta l}_0 k\mathrm{d}x = \dfrac{k\Delta l^2}{2}. 
\end{align}

<br>

Плотность же:

<br>

\begin{align}
    u = \dfrac{U}{V} = \dfrac{U}{Sl} = \dfrac{k\Delta l^2}{2Sl} = \boxed{\dfrac{E\varepsilon^2}{2}}.
\end{align}

<br>

### Коэффициент Пуассона

**Коэффициент Пуассона** - коэффициент пропорциональности между продольной и поперечной деформациями:

<br>

\begin{align}
    \varepsilon_a = \dfrac{\Delta a}{a} = -\mu \dfrac{\Delta l}{l} = \mu \varepsilon_l = -\mu \dfrac{\sigma}{E}.
\end{align}

<br>

Модуль Юнга и коэффициент Пуассона полностью характеризуют малые упругие деформации изотропных тел.

Малые упругие деформации подчиняются принципу суперпозиции.

### Деформация параллелепипеда

<br>

\begin{align}
    &\varepsilon_{x1} = \dfrac{\sigma}{E} \varepsilon_{xa} = -\mu\dfrac{\sigma_y}{E} - \mu \dfrac{\sigma_{z}}{E}\nonumber\\
    &\begin{cases}
        \varepsilon_x = \dfrac{\sigma_x}{E} - \dfrac{\mu}{E}\left(\sigma_y+\sigma_z\right)\nonumber\\
        \nonumber\\
        \varepsilon_y = \dfrac{\sigma_y}{E} - \dfrac{\mu}{E}\left(\sigma_x+\sigma_z\right)\nonumber\\
        \nonumber\\
        \varepsilon_z = \dfrac{\sigma_z}{E} - \dfrac{\mu}{E}\left(\sigma_x+\sigma_y\right).\nonumber\\
    \end{cases}
\end{align}

<br>

Запасённая энергия:

<br>

\begin{align}
    &U = \dfrac{1}{2}\underbrace{S_x\sigma_x}_{F_x}\Delta x\dfrac{l_x}{l_x} + \dfrac{1}{2} S_y\sigma_y\Delta y\dfrac{l_y}{l_y} + \dfrac{1}{2}S_z\sigma_z\Delta z = \nonumber\\
    &= \dfrac{V}{2}\left(\sigma_x\varepsilon_x+\sigma_y\varepsilon_y+\sigma_z\varepsilon_z\right).
\end{align}

<br>

Плотность энергии:

<br>

\begin{align}
    u = \dfrac{U}{V} = \dfrac{1}{2}\left(\sigma_x\varepsilon_x+\sigma_y\varepsilon_y+\sigma_z\varepsilon_z\right).
\end{align}

<br>

Рассмотрим случай сжатия параллелепипеда. Пусть он со всех сторон испытывает одинаковое давление: $\sigma_x=\sigma_y=\sigma_z=-P$, тогда

<br>

\begin{align}
    &\varepsilon_x = -\dfrac{P}{E} + \dfrac{\mu}{E} 2P = -\dfrac{P}{E}\left(1-2\mu\right)\\
    &\varepsilon_x = \varepsilon_y = \varepsilon_z = -\dfrac{P}{E}\left(1-2\mu\right).
\end{align}

<br>

А чему будет равно изменение объёма?

<br>

\begin{align}
    &V + \Delta V = \left(l_x+\Delta l_x\right)\left(l_y+\Delta l_y\right)\left(l_z+\Delta l_z\right)\\
    &V = l_xl_yl_z\\
    &\implies \Delta V =l_x\Delta l_y l_z + l_y\Delta l_x l_z + l_x \Delta l_z l_y + o\left(\Delta l\right) =\nonumber\\
    & = \varepsilon_y V + \varepsilon_x V + \varepsilon_z V = 3V\varepsilon = -\dfrac{3VP}{E}\left(1 - 2\mu\right)\nonumber\\
    & \boxed{\Delta V = - \dfrac{3VP}{E}\left(1-2\mu\right)}.\\
    & \boxed{\varepsilon_V = \dfrac{\Delta V}{V} = -\dfrac{3P}{E}\left(1-2\mu\right) = -\dfrac{P}{K},\;\;K = \dfrac{E}{3\left(1-2\mu\right)}}.
\end{align}

<br>

Соотношение справедливо не только для параллелепипеда, но и для любого тела, на которое со всех сторон действует одинаковое давление.

<br>

$E$ - модуль Юнга. 
Помним, что $F = k \Delta l$\
\begin{align}
    \dfrac{\Delta l}{l} = \dfrac{\sigma}{E} = \dfrac{F}{ES} \implies k \Delta l = E s \dfrac{\Delta l}{l}.
\end{align}

<br>

### Плотность энергии упругой деформации.

Энергия упругой деформации формируется за счёт работы силы по растяжению (сжатию) стержня:

<br>

\begin{align}
    &U = A = \int\limits_0^{\Delta l}F(x)dx = \int\limits_0^{\Delta l} kxdx=\dfrac{k\Delta l^2}{2}\nonumber \\
    &=\boxed{\dfrac{F\Delta l}{2}.}
\end{align}

<br>

Плотность энергии:

<br>

\begin{align}
    &u = \dfrac{U}{V} = \dfrac{U}{sl} = \dfrac{k\Delta l^2}{2sl} = \dfrac{k\Delta l}{2s}\cdot \varepsilon = \dfrac{E\varepsilon^2}{2}.\nonumber \\
    &\boxed{u=\dfrac{E\varepsilon^2}{2}}
\end{align}

<br>

### Коэффициент Пуассона.
Коэффициент пропорциональности между продольными $\Delta l/l$ и поперечными $\Delta a/a$ удлинениями.

<br>

\begin{align}
    \varepsilon_a = \dfrac{\Delta a }{a} = - \mu \dfrac{\Delta l}{l} = - \mu \varepsilon_l = \mu \dfrac{\sigma}{E}.
\end{align}

<br>

Чему бывает равна плотность энергии упругой деформации?

<br>

\end{align}

<br>

Обычно $0<\mu<1/2$

## Элементы теории волн
### Волновое уравнение
Мы рассматривали колебательный процесс. У колебаний есть фаза. Если разные части системы имеют разную фазу, то получаются волны. Пусть в нулевой момент времени волна описывается $f(x)$. В простейшем случае данная форма двигается в пространстве. Пусть в пространстве и времени такая волна описывается $S(x,t)$, где $S(x,0) = f(x)$.
Если форма не меняется, то за время $t$ происходит перемещение волны вправо, следовательно

<br>

\begin{align}
    S(x,t) = f(x - vt).
\end{align}

<br>

Волна, бегущая в положительном направлении $Ox$.
Смещение точек - $f$ -  постоянно для координаты-времени, $x - vt = const$.

<br>

\begin{align}
    &\implies \dfrac{d}{{d}t}(x - vt) = 0 \implies \dot{x} = v\dot{t} = v.
\end{align}

<br>

$\implies v$ - скорость постоянного значения величины $f$, то есть скорость волны.
Найдём уравнение волны, которому удовлетворяет данное решение.

<br>

\begin{align}
    &\dfrac{\partial}{\partial x}f = \dfrac{\partial f}{\partial u} \cdot \dfrac{\partial u}{\partial x} = f'\implies \dfrac{\partial^2f}{\partial x^2} = f''\nonumber \\
    &\dfrac{\partial }{\partial t} f = \dfrac{\partial f}{\partial u} \dfrac{\partial u }{\partial t} = f'(-v) \implies \dfrac{\partial^2 f}{\partial t^2} = v^2 f'', \\
    &\dfrac{\partial^2 f}{\partial x^2} = \dfrac{1}{v^2} \dfrac{\partial^2 f}{\partial t^2}. \\
    &\boxed{\dfrac{\partial^2 f}{\partial x^2} - \dfrac{1}{v^2} \dfrac{\partial^2 f}{\partial t^2} = 0}.
\end{align}

<br>

Полученное уравнение называют волновым уравнением, и иногда записывается с использованием даламбертиана $\square$.

<br>

\begin{align}
    &\square f \overset{\text{def}}{=} \dfrac{\partial^2 f}{\partial x^2} - \dfrac{1}{v^2} \dfrac{\partial^2 f}{\partial t^2}. \\
    &\boxed{\square f = 0}.
\end{align}

<br>

Более того, похожими рассуждениями можно обобщить волновое уравнение и на трёхмерный случай. Вместо второй частной производной по координате нам придётся взять сумму вторых частных производных по всем координатам, то есть дивергенцию градиента ―

<br>

\begin{align}
    &\nabla^2 f = \sum_j \dfrac{\partial^2 f}{\partial x_j^2}, \\
    &\boxed{\nabla^2 f - \dfrac{\partial^2 f}{\partial t^2} = 0}. \label{eq:wave_equation_3d}\\
    &\square f = \nabla^2 f - \dfrac{\partial^2 f}{\partial t^2}, \\
    &\square f = 0.
\end{align}

<br>

Зачем нам все эти рассуждения? Дело в том, что теперь каждый раз, когда нам случайно встретится волновое уравнение, мы сможем тут же без лишних рассуждений указать на его решение. Не важно, из чего мы вывели волновое уравнение для конкретного случая ― второго закона Ньютона, уравнений Максвелла, гидродинамических законов, каких-то термодинамических моделей, ― выполнение волнового уравнения гарантирует нам, что и струна, и свет, и круги на поверхности жидкости, и звук будут вести себя абсолютно одинаковым образом. 

Сейчас мы рассмотрим частные решения волнового уравнения, тем самым заочно решив немало физических задач.

### Плоская волна
Одним из решений волнового уравнения является плоская одномерная волна,

<br>

\begin{align}
    &f(x, t) = a_0 \cos (k (x - v t) + \varphi_0) \\
    &= a_0 \cos(k x - k v t + \varphi_0).
\end{align}

<br>

$k$ называют волновым числом, а $v$ ― скоростью волны (из соображений размерности: эта величина измеряется в метрах в секунду). Заметим, что это решение весьма похоже на уже рассмотренное ранее решение уравнения колебания ― по сути, волна и есть колебание, которое при этом как-то распространяется в пространстве. Отождествим величину $k v$ с коэффициентом при $t$ в решении уравнения колебания, то есть, с угловой частотой $\omega$. 

<br>

\begin{align}
    &k v = \omega, \\
    &k v = 2 \pi \nu, \\
    &k = 2 \pi \dfrac{\nu}{v} = 2 \pi \dfrac{1}{v T}.
\end{align}

<br>

Величина $v T$ имеет размерность длины и измеряется в метрах ― назовём её длиной волны $\lambda$.

<br>

\begin{align}
    &k = \dfrac{2 \pi}{\lambda}. \\
    &f(x, t) = a_0 \cos (k x - \omega t + \varphi_0). 
\end{align}

<br>

Довольно интересно, что период косинуса $2 \pi$, а значит, если зафиксировать некий момент времени и посмотреть на зависимость функции от координаты, то она будет периодичной с периодом $\dfrac{2 \pi}{k} = \lambda$ ― в этом и состоит физический смысл длины волны.

### Поверхности постоянной фазы
Обобщим случай одномерной волны на трёхмерное пространство. Рассмотрим решение волнового уравнения (в качестве упражнения можно подставить эту функцию в \ref{eq:wave_equation_3d} и проверить, что она и правда является решением)

<br>

\begin{align}
    &f(\vec{r}, t) = a_0 \cos (\vec{k} \cdot \vec{r} - \omega t + \varphi_0) = a_0 \cos(\varphi(\vec{r}, t)), \\
    &\varphi(\vec{r}, t) = (\vec{k} \cdot \vec{r}) - \omega t + \varphi_0.
\end{align}

<br>

Вектор $\vec{k}$ обобщает волновое число и называется волновым вектором, функцию $\varphi(\vec{r}, t)$ называют фазой. Рассмотрим поверхность, заданную как $\varphi(\vec{r}, t) = \text{const}$.

<br>

\begin{align}
    &(\vec{k} \cdot \vec{r}) = k_x \cdot x + k_y \cdot y + k_z \cdot z. \\
    &\varphi = k_x x + k_y y + k_z z - \omega t + \varphi_0 = \text{const}. \\
    &\underbrace{k_x}_A x + \underbrace{k_y}_B y + \underbrace{k_z}_C z + \underbrace{- \omega t + \varphi_0 - \varphi}_{D} = 0. \\
    &A x + B y + C z + D = 0.
\end{align}

<br>

Весьма скромных познаний в аналитической геометрии достаточно, чтобы узнать в последнем выражении уравнение плоскости в трёхмерном пространстве. Интересно, что вектор нормали $\vec{n} = \begin{pmatrix} A \\ B \\ C \end{pmatrix} = \begin{pmatrix} k_x \\ k_y \\ k_z \end{pmatrix} = \vec{k}$, то есть волновой вектор перпендикулярен поверхности одинаковой фазы и показывает направление распространения волны. Величина $D$ в данном случае определяет, точки с какой конкретно фазой нас интересуют. Интересно, что она зависит от времени.

### Пригоршня других решений волнового уравнения
Решениями волнового уравнения являются и другие интересные функции, например

<br>

\begin{align}
    &f(\vec{r}, t) = \dfrac{a_0 \cos (\vec{k} \cdot \vec{r} - \omega t + \varphi_0)}{r} \text{ ― сферические волны}, \\
    &f(\vec{r}, t) = \dfrac{a_0 \cos (\vec{k} \cdot \vec{r} - \omega t + \varphi_0 }{\sqrt{r}} \text{ ― цилиндрические волны}.
\end{align}

<br>

Более того, волна не обязательно должна задаваться некой скалярной функцией ― ничто не мешает рассмотреть и векторную $\vec{f}(\vec{r}, t) = \vec{f}(\vec{k} \cdot \vec{r} - \omega t)$. В таком случае, если $\vec{f}$ параллельна распространению волны, то такие называют продольными, а если $\vec{f}$ перпендикулярна, то, наоборот, поперечными.


### Принцип суперпозиции
Очень важно для понимания волн помнить, что для любых двух решений волнового уравнения $f_1(x, t)$ и $f_2(x, t)$ их линейная комбинация $\alpha f_1 + \beta f_2$ будет также являться решением волнового уравнения. На языке даламбертианов это можно записать как $\square(\alpha f_1 + \beta f_2) = \alpha \square f_1 + \beta \square f_2$. (Если читатель знаком с теорией линейных операторов, то он может заметить, что изучение свойств даламбертиана может привести к интересным выводам о физической природе самых разных волновых процессов.)

### Стоячие волны
Складывая две волны одинаковой амплитуды и частоты, бегущие навстречу друг другу с одинаковой скоростью, мы получаем очень интересное явление ― стоячие волны.

<br>

\begin{align}
    &a_0 \cos (k x - \omega t) + a_0 \cos (- k x - \omega t) = 2 a_0 \cos (k x) \cos (\omega t), \\
    &f(x, t) = \underbrace{2 a_0 \cos (kx)}_A \cos (\omega t) = A(x) \cos(\omega t).
\end{align}

<br>

Таким образом, мы получим обычные колебания в каждой точке, амплитуда которых будет изменяться с координатой. Точки с нулевой амплитудой называют узлами ― в них колебание вырождается и превращается в тождественный ноль. Точки с максимальной амплитудой, где $\cos (k x) = \pm 1$, называют пучностями. 

Из периодичности косинуса, расстояние между ближайшими точками с одинаковой по модулю амплитудой (в том числе, расстояние между двумя соседними узлами или расстояние между двумя соседними пучностями) всегда одинаково и в точности равняется половине длины волны \dfrac{\lambda}{2}.
