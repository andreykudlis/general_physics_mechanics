# Колебания. Продолжение

Через какое-то время решение однородного уравнения - затухающее решение полностью исчезнет. Остается только частное решение.

Амплитуда зависит от частоты. Есть максимальное значение этой амплитуды, т.е. мы получаем резонанс - явление резкого возрастания амплитуды.

Анализ $C(\Omega)$:

<br>



1) При $\Omega \approx 0 \Rightarrow |C|\;=\;\dfrac{f_0}{\omega_0^2}$ и $\psi < 0$.

2) При $\Omega \gg \omega_0 \Rightarrow |C|\;=\; \dfrac{f_0}{\Omega \sqrt{\Omega^2 + 4 \gamma^2}} \approx 0$ и $\psi \rightarrow 0$.

3) $\max{\dfrac{d|C|}{d\Omega}}\;=\;0\;=\;-\dfrac{1}{2} \dfrac{f_0 \cdot (2(\omega_0^2 - \Omega^2)2 \Omega (-1) + 8\gamma^2 \Omega)}{((\omega_0^2 - \Omega^2)^2 + 4\gamma^2 \Omega^2)^{\frac{3}{2}}} = 0$

<br>


\begin{align*}
    &\Rightarrow -4\Omega  (\omega_0^2 - \Omega^2) + 8\gamma^2 \Omega = 0,\\
    &2\gamma^2 = \omega_0^2 - \Omega^2,\\
    &\Omega = \sqrt{\omega_0^2 - 2\gamma^2}.
\end{align*}

<br>

\begin{align*}
    &|C_{res}| = \dfrac{f_0}{\sqrt{4\gamma^4 + 4\gamma^4(\omega_0^2 - 2\gamma^2)}} = \dfrac{f_0}{2\gamma \sqrt{\gamma^2 + \omega_0^2 - 2\gamma^2}} = \\
    &= \dfrac{f_0}{2\gamma \sqrt{\omega_0^2 - \gamma^2}}.
\end{align*}

<br>

При малых $\gamma$: $\Omega \approx \omega_0$

<br>

Фаза в резонансе:

<br>

\begin{align}
    &\tan{\varphi} = - \dfrac{2\gamma \sqrt{\omega_0^2 - 2\gamma^2}}{2\gamma^2}\;=\; - \dfrac{\sqrt{\omega_0^2 - 2 \gamma^2}}{\gamma} .
\end{align}

<br>

**Отношение амплитуды в резонансе к статистическому сдвигу:**

<br>

\begin{align}
    &\dfrac{|C_{res}|}{\dfrac{f_0}{\omega_0^2}}\;=\; \dfrac{\omega_0^2}{2\gamma \sqrt{\omega_0^2\;-\; \gamma^2}} \; \approx \; \dfrac{\omega_0}{2\gamma}\;=\; \dfrac{\pi}{\gamma T} .
\end{align}

<br>

- добротность.

Найдем ширину резонансного пика: по определению это ширина опред. значением частоты, где $|C|\;=\; \dfrac{|C_{res}|}{\sqrt{2}}$ .

<br>

\begin{align}
    &\dfrac{f_0}{\sqrt{(\omega_0^2 - \Omega_{1,2}^2)^2 + 4\gamma^2 \Omega^2}}\;=\; \dfrac{1}{\sqrt{2}} \dfrac{f_0}{2\gamma \sqrt{\omega_0^2 - \gamma^2}},\\
    &\Rightarrow (\omega_0^2 - \Omega_{1,2}^2)^2 \;=\; 2 \cdot 4\gamma^2 \omega_0^2 - 4\gamma^2 \Omega_{1,2}^2,\\
    &(\omega_0^2 - \Omega_{1,2}^2)^2 \cong 4\gamma^2 \omega_0^2,\\
    &\Rightarrow \Omega_{1,2} \approx +\sqrt{\omega_0^2 \pm 2\gamma \omega_0}\;=\; \omega_0 \sqrt{1 \pm \dfrac{2\gamma}{\omega_0}}\approx \omega_0\pm\gamma,\\
    &\Rightarrow \dfrac{\Omega_{res}}{\Delta \Omega}\; \cong \; \dfrac{\omega_0}{2\gamma}\;=\; \dfrac{\pi}{T \gamma}.
\end{align}

<br>

## Установление колебаний

Переходный процесс - процесс установления вынужденных колебаний. 

Характерное время такого процесса: $\tau = \dfrac{1}{\gamma}$. За это время амплитуда затухающих колебаний уменьшается в $e$ раз

Посмотрим на установление такого режима в случае раскачки гармонического осциллятора со слабым затуханием, с частотой, которая примерно равна частоте собственных колебаний.

<br>


$$x(t)\;=\; \dfrac{f_0}{\sqrt{(\omega_0^2 - \Omega^2)^2 + 4\gamma^2 \Omega^2}}\cos{(\Omega t + \psi)} + ae^{-\gamma t} \cos({\omega t + \varphi_0})$$

<br>


Пусть $x_0 = \dfrac{f_0}{\omega_0^2}$

<br>


\begin{align}
    &x(t) = \dfrac{\omega_0}{2\gamma} x_0 \cos{(\omega_0 t+ \psi)} + ae^{-\gamma t} \cos({\omega t + \varphi_0}), \\
    &\tan{\psi}\;=\; - \dfrac{2\gamma \Omega}{\omega_0^2 - \Omega^2} \rightarrow - \infty \Rightarrow \psi \;=\; - \dfrac{\pi}{2}, \\
    &x(t) = Qx_0 \sin{\omega_0 t} + ae^{-\gamma t} \cos({\omega t + \varphi_0}), \\
    &\dfrac{dx}{dt} \approx Qx_0 \cos{(\omega_0 t)} \omega_0 - ae^{-\gamma t} \sin({\omega t + \varphi_0}) \omega_0.
\end{align}

<br>


Начинаем раскачивать: $x(0) = 0, \dfrac{dx}{dt}(0) = 0$

<br>

$\varphi_0 \;=\; \dfrac{\pi}{2}$, $a\;=\; Qx_0$
$$x(t)\;=\; Qx_0 \sin{\omega_0t}[1\;-\; e^{-\gamma t}]$$

## Мощность затухания

Рассмотрим установившийся режим:

<br>

\begin{align}
    &P_{\text{тр}}\;=\; \dfrac{\delta A}{dt}\;=\; \dfrac{F_{\text{тр}} dx}{dt}\;=\; -\beta (\dfrac{dx}{dt})^2\;=\\
    &=\; \beta \left[ \dfrac{d}{dt} |C| \cos({\Omega t + \varphi}) \right]^2\;=\; -\beta |C|^2 \Omega^2 \sin^2({\Omega t + \varphi})\;=\\
    &=\; -2\gamma m \Omega^2 |C|^2 \sin^2({\Omega t + \varphi}).
\end{align}

<br>


Усредним по периоду:

<br>


\begin{align}
    <P_\text{тр}> = \dfrac{1}{T} \cdot \underset{o}{\overset{T}{\int}} P_\text{тр} (t) dt = - \Omega^2 \cdot \gamma m |c|^2 = -\dfrac{\Omega \gamma m  \cdot f_0 ^2}{(\Omega^2 - \omega_0 ^2)^2 + 4\gamma^2 \Omega^2}
\end{align}

<br>


## Мощность внешней силы

<br>

\begin{align}
    & \nonumber P_F = mf_0 cos(\Omega t) \cdot \dfrac{dx}{dt} = mf_0 cos(\Omega t) \cdot (-1) |c| sin (\Omega t + \psi)\Omega =
    \\
    & \nonumber = - mf_0 |c| \Omega cos(\Omega t) \cdot  sin (\Omega t + \psi)  = \\ 
    &= - mf_0 |c| \Omega [ cos\Omega t \cdot sin\Omega t \cdot cos\psi + cos\Omega t \cdot cos\Omega t + sin\psi ] =
    \\
    & = - mf_0 |c| \Omega [ cos\Omega t \cdot sin\Omega t \cdot cos\psi + cos^2\Omega t + sin\psi ] =\\ 
    &= - mf_0 |c| \Omega [ cos\psi \dfrac{sin2\Omega t}{2} + cos^2 \Omega t + sin\psi ].
    \\
    & <P_F> = - \dfrac{mf_0 |c| \Omega}{2} \cdot sin\psi = -\dfrac{mf_0 f_0 \Omega sin\psi}{2\sqrt{(\Omega^2 - \omega_0 ^2)^2 + 4\gamma^2\Omega^2}} = 
    \\
    & = \dfrac{mf_0 ^2 \Omega}{\sqrt{(\Omega^2 - \omega_0 ^2)^2 + 4\gamma^2\Omega^2}} \cdot \dfrac{\gamma \Omega}{\sqrt{(- \Omega^2 + \omega_0 ^2)^2 + 4\gamma^2\Omega^2}}.
\end{align}

<br>

$$sin\psi = \dfrac{1}{\sqrt{1 + \dfrac{1}{tg^2 \psi}}},$$

$$tg\psi = -\dfrac{2\gamma \Omega}{\omega^2 - \Omega^2}.$$
