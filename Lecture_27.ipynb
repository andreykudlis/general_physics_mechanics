# Свободные колебания двух связанных осцилляторов

Сперва запишем самый общий случай:

<br>

\begin{align}
    \begin{cases}
        & m_1 \ddot{x}_1 = -m_1g\varphi_1 - \beta_1 \dot{x}_1 - k(x_1-x_2)
        \\
        & m_2 \ddot{x}_2 = -m_2g\varphi_2 - \beta_2 \dot{x}_2 - k(x_2-x_1)
    \end{cases}
    \\
    \begin{cases}
        & m_1 \ddot{x}_1 = -m_1g\dfrac{x_1}{l} - \beta_1 \dot{x}_1 - k(x_1-x_2)
        \\
        & m_2 \ddot{x}_2 = -m_2g\dfrac{x_2}{l} - \beta_2 \dot{x}_2 - k(x_2-x_1)
    \end{cases}
\end{align}

<br>

Данная система имеет аналитическое решение, однако анализировать его нет смысла (очень большое).

Рассмотрим случай малых затуханий:
Пусть $\beta_1$ и $\beta_2 << \sqrt{km}$

<br>

\begin{align}
    & \begin{cases}
        & \dfrac{m_1 \ddot{x}_1 + m_2\ddot{x}_2}{m_1 + m_2} = - \dfrac{m_1 x_1 + m_2 x_2}{m_1 + m_2} \cdot \dfrac{g}{l}
        \\
        & \ddot{x}_2 - \ddot{x}_1 = -\dfrac{g}{l} (x_2 - x_1) - \bigg( \dfrac{k}{m_2} + \dfrac{k}{m_1}\bigg) (x_2 - x_1)
    \end{cases}
    \\
    & \begin{cases}
        & \ddot{u} + \dfrac{g}{l}u = 0
        \\
        & \triangle \ddot{x} + (\dfrac{g}{l} + \dfrac{k}{\mu}) \triangle x = 0
    \end{cases}
\end{align}

<br>

\begin{align}
    &\implies u =u_0\cos(\omega_ut+\varphi_u),\nonumber \\
    &\Delta x =\Delta x_0\cos(\omega_{\Delta x}t + \varphi_{\Delta x}).\nonumber \\
    &u = \dfrac{m_1x_1 + m_2x_2}{m_1+m_2}, \ \  \Delta x = x_2 - x_1.
\end{align}

<br>

Понятно, что $\omega_{\Delta x} > \omega_u$

<br>

\begin{align}
    &\implies x_1 = u - \dfrac{\Delta xm_2}{m_1 + m_2}; \nonumber \\
    &\implies x_2 = u + \dfrac{\Delta xm_1}{m_1 + m_2}.
\end{align}

<br>

Подставим полученные ранее выражения для $\Delta x$ и $u$:

<br>

\begin{align}
    &x_1 = u\cos(\omega_u t + \varphi_u) - \dfrac{m_2}{m_1 + m_2} \Delta x_0\cos(\omega_{\Delta x}t + \varphi_{\Delta x}),\nonumber \\
    &x_2 = u\cos(\omega_u t + \varphi_u) + \dfrac{m_1}{m_1 + m_2} \Delta x_0\cos(\omega_{\Delta x}t + \varphi_{\Delta x}).
\end{align}

<br>

Рассмотрим пример: пусть $x_1(0) = 0; \ x_2(0) = x_0; \ v_1(0) = v_2(0) = 0.$
$ \ m_1 = m_2 = m,  \mu = m^2/2m = m/2.$

<br>

\begin{align}
    &\begin{cases}
        &x_1 = u_0\cos(\omega_u t + \varphi_u) - \dfrac{\Delta x_0}{2} \cos(\omega_{\Delta x}t + \varphi_{\Delta x}), \\
        &x_2 = u_0\cos(\omega_u t + \varphi_u) + \dfrac{\Delta x_0}{2} \Delta x_0\cos(\omega_{\Delta x}t + \varphi_{\Delta x}).
    \end{cases}\\
    &\begin{cases}
        &u_0\cos(\varphi_u) - \dfrac{\Delta x_0}{2} \cos(\varphi_{\Delta x}) = 0, \\
        &u_0\cos(\varphi_u) + \dfrac{\Delta x_0}{2} \cos(\varphi_{\Delta x}) = x_0,\\
        &-u_0\omega_u\sin(\varphi_u) + \dfrac{\Delta x_0}{2}\omega_{\Delta x} \sin(\varphi_{\Delta x}) = 0,\\
        &-u_0\omega_u\sin(\varphi_u) - \dfrac{\Delta x_0}{2}\omega_{\Delta x} \sin(\varphi_{\Delta x}) = 0.
    \end{cases}
\end{align}

<br>

Сложим и вычтем парами первые два и последние два уравнения системы:

<br>

\begin{align}
    &\begin{cases}
     &2u_0\cos\varphi_u = x_0,\\
     &\Delta x_0 \cos\varphi_{\Delta x} = x_0,\\
     &u_0\sin\varphi_u = 0,\\
     &\Delta x_0 \sin\varphi_{\Delta x} = 0.
    \end{cases}
\end{align}

<br>

Из первого и третьего уравнения новой системы следует, что $\varphi_u = 0$, а из второго и четвёртого следует, что $\varphi_{\Delta x}=0$, и тогда:

<br>

\begin{align}
    &u_0 = \dfrac{x_0}{2},\\
    &\Delta x_0 = x_0.
\end{align}

<br>

Зная значения $u_0$ и $\Delta x_0$ знаем, что:

<br>

\begin{align}
    &\begin{cases}
            &x_1 = \dfrac{\Delta x_0}{2}\left[\cos(\omega_u t) -  \cos(\omega_{\Delta x}t)\right], 
            &x_2 = \dfrac{\Delta x_0}{2}\left[\cos(\omega_u t) +  \cos(\omega_{\Delta x}t)\right].
    \end{cases}\\
    \implies 
    &\begin{cases}
            &x_1 = \dfrac{x_0}{2}\left[-2\sin\left(\dfrac{\omega_u-\omega_{\Delta x}}{2} t\right) \sin\left(\dfrac{\omega_u+\omega_{\Delta x}}{2} t\right)\right], 
            &x_2 = \dfrac{x_0}{2}\left[2\cos\left(\dfrac{\omega_u + \omega_{\Delta x}}{2} t\right) \cos\left(\dfrac{\omega_u-\omega_{\Delta x}}{2} t\right)\right].
    \end{cases}
\end{align}

<br>

Добавим внешнюю вынуждающую силу и рассмотрим установившиеся колебания:

<br>

\begin{align}
    &\begin{cases}
        &\ddot{x}_1 + \omega_0^2x_1 + 2\gamma_1\dot{x}_1 + \dfrac{k}{m_1}(x_1-x_2) = f_0 \cos(\Omega t),
        &\ddot{x}_2 + \omega_0^2x_2 + 2\gamma_2\dot{x}_2 + \dfrac{k}{m_2}(x_2-x_1) = 0.
    \end{cases}
    &\begin{cases}
        &\ddot{x}_1 + \left(\omega_0^2 + \dfrac{k}{m_1}\right) x_1 + 2\gamma_1\dot{x}_1 - \nu_1x_2 = f_0 \cos(\Omega t),\\
        &\ddot{x}_2 + \left(\omega_0^2 + \dfrac{k}{m_2}\right) x_2 + 2\gamma_2\dot{x}_2 - \nu_2x_1 = 0.
    \end{cases}
\end{align}

<br>

$\nu_1 = k/m_1$ и $\nu_2 = k/m_2$,  $\omega_0^2 + k/m_1 = \omega_{01}^2, \omega_0^2 + k/m_2 = \omega_{02}^2$.
$x_1 =C_1e^{i\Omega t},  x_2 = C_2e^{i\Omega t}$.

<br>

\begin{align}
    &\begin{cases}
        &-\Omega^2C_1 + \omega_{01}^2C_1 + 2\gamma_1 C_1 i \Omega - \nu_1 C_2 = f_0,\\
        &-\Omega^2C_2 + \omega_{02}^2C_2 + 2\gamma_2 C_2i\Omega - \nu_2 C_1 = 0.
     \end{cases}
     \\
    &\begin{cases}
        &C_1\underbrace{(-\Omega^2 + \omega_{01}^2 + 2\gamma_1 i\Omega)}_{A_1} \underbrace{- \nu_1}_{B_1} C_2 = f_0,\\
        &C_2\underbrace{(-\Omega^2 + \omega_{02}^2 + 2\gamma_2 i\Omega)}_{A_2} \underbrace{- \nu_2}_{B_2} C_1 = 0.
     \end{cases}
\end{align}

<br>

\begin{align}
    &\begin{cases}
        &C_1 A_1 + B_1 C_2 = f_0,\\
        &C_2 A_2 + B_2 C_1 = 0.
     \end{cases}
     \\
    &\begin{cases}
        &C_2 = - \dfrac{B_2}{A_2}C_1, \\
        &C_1 A_1 + B_1 \left(-\dfrac{B_2}{A_2}C_1\right) = f_0.
    \end{cases}
    \\
    &\begin{cases}
        &C_2 = -\dfrac{B_2}{A_2}C_1 = -\dfrac{B_2}{A_2} \dfrac{f_0}{A_1 - \dfrac{B_1B_2}{A_2}},\\
        &C_1\left(A_1- \dfrac{B_1 B_2}{A_2}\right) = f_0.
    \end{cases}
    \\
    &\begin{cases}
        &C_1 = \dfrac{f_0A_2}{A_1A_2 - B_1B_2},\\
        &C_2 = \dfrac{-f_0B_2}{A_1A_2 - B_1 B_2}.
    \end{cases}
 \end{align}
 
 <br>
 
 Теперь посмотрим на явление резонанса. Для этого посмотрим на знаменатель:
 
 <br>
 
 \begin{align}
     &A_!A_2 - B_1B_2 = (\omega_{01}^2 - \Omega^2 + 2\gamma_1i\Omega)(\omega_{02}^2 - \Omega^2 + 2\gamma_2 i \Omega) - \nu_1\nu_2 \nonumber \\
     &= \Omega^4 - (\omega_{01}^2+\omega_{02}^2)\Omega^2 + \omega_{01}^2 \omega_{02}^2 - \nu_1\nu_2 + 2\gamma_2 i \Omega \omega_{01}^2 - \\
     &\Omega^32\gamma_2 i + 2\gamma_1 i \Omega\omega_{02}^2  - \Omega^3 2\gamma_1 i -4\gamma_1\gamma_2\Omega^2.
\end{align}
Рассмотрим нулевой порядок по $\gamma_{1,2}$:
\begin{align}
    &\Omega_{1, 2}^2 = \dfrac{\omega_{01}^2 + \omega_{02}^2 \pm \sqrt{(\omega_{01}^2 + \omega_{02}^2)^2 - 4 (\omega_{01}^2 + \omega_{02}^2 - \nu_1 \nu_2 )}}{2} \nonumber \\
    &= \dfrac{\omega_{01}^2 + \omega_{02}^2 \pm \sqrt{(\omega_{01}^2 - \omega_{02}^2)^2 + 4 \nu_1 \nu_2}}{2} \nonumber \\
    &= \dfrac{\omega_{01}^2 + \omega_{02}^2 \pm (\omega_{01}^2 - \omega_{02}^2) \sqrt{1 + \dfrac{4 \nu_1 \nu_2}{(\omega_{01}^2 - \omega_{02}^2)^2}}}{2}.
\end{align}

<br>

При малых колебаниях можем использовать приближение $(1 + \xi)^\alpha = 1 + \alpha \xi$, в нашем случае $\alpha = \dfrac{1}{2}$.

<br>

\begin{align}
    &\Omega_{1, 2}^2 \approx \dfrac{\omega_{01}^2 + \omega_{02}^2 \pm (\omega_{01}^2 - \omega_{02}^2) \left(1 + \dfrac{2 \nu_1 \nu_2}{(\omega_{01}^2 - \omega_{02}^2)^2}\right)}{2}, \\
    &\Omega_1^2 = \omega_{01}^2 + \dfrac{\nu_1 \nu_2}{\omega_{01}^2 - \omega_{02^2}}, \\
    &\Omega_2^2 = \omega_{02}^2 - \dfrac{\nu_1 \nu_2}{\omega_{01}^2 - \omega_{02^2}}. \\
    &C_1 = \dfrac{f_0 A_2}{(\Omega^2 - \Omega_1^2) (\Omega^2 - \Omega_2^2)} = \dfrac{a_1}{\Omega^2 - \Omega_1^2} + \dfrac{a_2}{\Omega^2 - \Omega_2^2}, \\
    &C_2 = \dfrac{f_0 B_2}{(\Omega^2 - \Omega_1^2) (\Omega^2 - \Omega_2^2)} = \dfrac{b_1}{\Omega^2 - \Omega_1^2} + \dfrac{b_2}{\Omega^2 - \Omega_2^2}.
\end{align}

<br>

То есть, при малых затуханиях и не очень сильной связи отклик элементов это в точности сумма откликов двух осцилляторов.

## Параметрическое возбуждение колебаний

Параметрические колебания ― такие, при которых увеличивается амплитуда за счёт периодического изменения какого-либо параметра системы.

Для примера рассмотрим математический маятник, который будет увеличивать свою амплитуду за счёт уменьшения длины.

<br>


В качестве начальных условий нам известны $\varphi_0$ и $l_0$. Посмотрим на процессы между событиями на рисунке: \\
$(0) \to (1)$: из сохранения энергии 

<br>

\begin{align}
    &mgl_0 (1 - \cos \varphi_0) = \dfrac{1}{2} m v_0^2.
\end{align}

<br>

$(1) \to (2)$: из сохранения момента импульса (сила направлена вдоль верёвки)

<br>

\begin{align}
    &m v_0 l_0 = m v_1 (l_0 - \Delta l). \\
    &\left\{ \begin{array}{l}
        v_1 = v_0 \dfrac{l_0}{l_0 - \Delta l} = \dfrac{2 g l_0}{l_0 \Delta l} \sqrt{1 - \cos \varphi_0}, \\ \\
        v_0 = \sqrt{2 g l_0 (1 - \cos \varphi_0}).
    \end{array} \right.
\end{align}

<br>

$(2) \to (3)$: из сохранения энергии

<br>

\begin{align}
    &\dfrac{1}{2} mv_1^2 = m g (l - \Delta l) (1 - \cos \varphi_1).
\end{align}

<br>

$(3) \to (4)$: из сохранения момента импульса скорость в этих точках нулевая, что, однако, автоматически выполняется. 

Затем на новом витке всё повторится заново, но уже с другим начальным отклонением, $\varphi_1$.

Выведем аналитическое выражение для $\varphi_j$, считая его много меньше единицы, а также принимая изменение длины малым по сравнению с начальной длиной маятника.
$$\require{enclose}$$

<br>

\begin{align}
    &\dfrac{1}{2} mv_1^2 = m g (l_0 - \Delta l) (1 - \cos \varphi_1). \\
    &v_1 = \dfrac{\sqrt{2g l_0} l_0}{l_0 - \Delta l} \sqrt{1 - \cos \varphi_0}. \\
    &\dfrac{1}{\enclose{updiagonalstrike}{2}} \dfrac{\enclose{updiagonalstrike}{g} l_0^3}{(l_0 - \Delta l)^2} \left( \sqrt{\dfrac{2 \varphi_0}{2}} \right)^2 \approx {\enclose{updiagonalstrike}{g}} (l_0 - \Delta l) \left(1 - \left( 1 - \dfrac{\varphi_1^2}{2} \right) \right), \\
    &\dfrac{l_0^3}{(l_0 - \Delta l)^3} \dfrac{{\enclose{updiagonalstrike}{\sqrt{2}}} \varphi_0^2}{\enclose{updiagonalstrike}{\sqrt{2}}} = \varphi_1^2, \\
    &\varphi_1 = \left(\dfrac{l_0}{l_0 - \Delta l}\right)^{\dfrac{3}{2}} \varphi_0. \\
    &\varphi_1 = \left( \dfrac{1}{1 - \dfrac{\Delta l}{l_0}} \right)^{\dfrac{3}{2}} \varphi_0 \approx \left(1 + \dfrac{3}{2} \dfrac{\Delta l}{l_0} \right) \varphi_0.
\end{align}

<br>
