# Адиабатические инварианты. Колебания АТТ. Упругость

## Адиабатические инварианты.

Под адиабатическими изменениями какого-либо параметра системы подразумевается медленное изменение. Если при этом медленном изменении параметра есть какая-либо комбинация физических характеристик системы, которая сохраняется, то такая комбинация называется адиабатическим инвариантом.

Рассмотрим пример:

Пусть есть система (материальная точка на пружине), которая описывается одномерными гармоническими колебаниями. Энергия таких колебаний, как нам хорошо известно, описывается формулой:

<br>

\begin{align}
    E = \dfrac{\alpha \dot{x}^2}{2} + \dfrac{\beta x^2}{2}.
\end{align}

<br>

Пусть $\beta$ меняется, но очень медленно (пружина растягивается неупруго):

<br>

\begin{align}
    &\dfrac{dE}{dt} = \dfrac{2\alpha \ddot{x}\dot{x}}{2} + \dfrac{2\beta x}{2}\dot{x} + \dfrac{x^2}{2}\dot{\beta},\nonumber \\
    &= \alpha\dot{x}\underbrace{\left(\ddot{x} + \dfrac{\beta}{\alpha}x\right)}_{=0} + \dfrac{\dot{\beta}x^2}{2}.\\
\end{align}

<br>

Таким образом:

<br>

\begin{align}
    &\dfrac{dE}{dt} = \dfrac{\dot{\beta}x^2}{2},\nonumber \\
    &E = \dfrac{\beta a^2}{2},\nonumber \\
    &x = a\cos(\omega_0t + \varphi_0). 
\end{align}

<br>

Рассмотрим изменение энергии:

<br>

\begin{align}
    &\Delta E = \int\limits_0^T dE = \dfrac{\dot{\beta}}{2}\int\limits_0^Ta^2cos^2(\omega_0 t + \varphi_0)dt,\nonumber \\
    &= \dfrac{\dot{\beta}}{2} a^2 \dfrac{1}{2}T = \dfrac{\Delta \beta}{2 T \beta} \dfrac{T\beta a^2}{2},\nonumber \\
    &\implies dE = \dfrac{d\beta E}{2\beta} \implies \dfrac{dE}{E} = \dfrac{1}{2} \dfrac{d\beta}{\beta},\\
    &\ln E = \ln \sqrt{\beta} + const,\nonumber \\
    & \dfrac{E}{\sqrt{\beta}} = const \implies \omega_0^2 = \dfrac{\beta}{\alpha} \implies \omega_0^2 \sim \beta \sim \dfrac{1}{T^2}.\nonumber \\
    &\boxed{E\cdot T = const.}
\end{align}
### Колебания твёрдого тела. Элементы теории упругости.
Физический маятник - твёрдое тело, которое может вращаться вокруг неподвижной оси.

В нашем случае можно сказать, что АТТ совершает малые колебания в окрестности равновесия.

Пусть полная энергия: $E = T + U$.

Мы помним, что в случае вращательного движения: $T = I\omega^2/2$, а потенциальная энергия пропорциональна квадрату координаты (в нашем случае углу отклонения): $U = \beta \varphi^2 /2$.

Запишем полную энергию:

<br>

\begin{align}
    &E = \dfrac{I\dot{\varphi}^2}{2} + \dfrac{\beta \varphi^2}{2}, \nonumber \\
    &\implies \dfrac{dE}{dt} = 0 \implies \dfrac{I 2 \dot{\varphi}}{2}\ddot{\varphi} + \dfrac{\beta 2 \varphi}{2}\dot{\varphi} = 0,\nonumber \\ 
    &\ddot{\varphi} + \dfrac{\beta}{I}\varphi = 0 \implies \Omega^2 = \dfrac{\beta}{I} = \dfrac{mga}{I}.
\end{align}

<br>

Для маятника $\beta = mga$, где $a$ - расстояние до центра масс.

Хотим записать потенциальную энергию. Чем она определяется? Положением центра масс стержня! То есть координатой z, либо введём h.
длина верёвки = const

<br>

\begin{align}
    &l^2 = (x_A - x_{B(0)})^2 + (y_A - y_{B(0)})^2 + (z_A - z_{B(0)})^2
    \\
    & l^2 = (x_A - x_{B(t)})^2 + (y_A - y_{B(t)})^2 + (z_A - z_{B(0)})^2
    \\
    &x_A= a
    \\
    &y_A = 0
    \\
    &z_A = 0
\end{align}

<br>

Если  $\varphi$ - угол отклонения от положения равновесия, то:

<br>

\begin{align}
    &x_{B(0)} = a, \; \; x_{B(t)} = a\cos \varphi
    \\
    &y_{B(0)} = 0, \; \; y_{B(t)} = a\sin\varphi
    \\
    &z_{B(0)} = -l, \; \; z_{B(t)} = -l + h
    \\
    & \Rightarrow l^2 = (acos\varphi - a)^2 + (asin\varphi)^2 + (-l+h)^2
    \\
    &l^2 = a^2cos^2\varphi - 2a^2cos\varphi + a^2 + a^2sin^2\varphi + l^2 + h^2 -2lh = 0
    \\
    & la^2(1 - \varphi) = h(2l - h).
    \end{align}

<br>

Рассмотрим малые колебания:

<br>

\begin{align}
    & \dfrac{a^2 \varphi^2}{2} = lh \; \Rightarrow h = \dfrac{a^2 \varphi^2}{2l}
    \\
    & U = mg \cdot \dfrac{a^2 \varphi^2}{2l}
    \\
    & E = \dfrac{I \dot{\varphi}^2}{2} + mg\dfrac{a^2 \varphi^2}{2l}
    \\
    & \Rightarrow \Omega^2 = \dfrac{mga^2}{lI} 
    \\
    & I \text{ - момент инерции относительно оси вращения.}
\end{align}

### Теорема Гюйгенса

Введём определения:

Приведённая длина физического маятника - это длина такого математического маятника, чей период совпадает с периодами физического маятника:

<br>

\begin{align}
    T = 2\pi \sqrt{\dfrac{l_{\text{пр}}}{g}} = 2\pi \sqrt{\dfrac{I}{mga}} \Rightarrow \boxed{l_{\text{пр}} = \dfrac{I}{ma}} .
\end{align}

<br>

Центр качаний - точка, где нужно сосредоточить массу маятника, чтобы период колебаний не изменился.

$\Rightarrow$  центр качания отстоит от оси подвеса на приведённую длину.
Замечание: центр качения находится всегда дальше центра масс:

<br>

\begin{align}
    l_{\text{пр}} = \dfrac{I}{ma} = \dfrac{I_c + ma^2}{ma} = a + \dfrac{I_c}{ma} .
\end{align}

<br>

**Теорема** Если подвесить маятник за центр качания, то период его колебаний не изменится, а прежняя точка подвеса станет новым центром качания: чему равно расстояние от центра масс до центра качания:

<br>

\begin{align}
    \dfrac{I_c}{ma} = b \Rightarrow l_{\text{пр}} = a + b
\end{align}

<br>

, если взять теперь за центр качания (сделать его новой точкой подвеса) $\Rightarrow b$ - расстояние до центра масс $\Rightarrow$ по определению $b$ входит в выражение для 

<br>

\begin{align}
    l_{\text{прив b}} = b + \dfrac{I_c}{mb} = \dfrac{I_c}{ma} + a = l_{\text{прив a}} \Rightarrow l_{\text{пр b}} = l_{\text{пр a}} = l_{\text{пр}}.
\end{align}
